```python
df = pd.read_csv("../input/iris-data/Iris.csv")
df
```

```python
target_column = df['Species']

# Recode 'Setosa' to 0
target_column = np.where(target_column == 'Iris-setosa', 0, target_column)
target_column = np.where(target_column == 'Iris-versicolor', 1, target_column)
target_column = np.where(target_column == 'Iris-virginica', 2, target_column)
# target_column

df['Species'] = target_column
df
```

```python
X = df
X = X.drop('Id', axis=1)
X

# Define the desired order of columns
desired_order = ['Species', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Rearrange the columns using reindex
X = X.reindex(columns=desired_order)
X
```

```python
first_col = X.iloc[:, 0]
other_cols = X.iloc[:, 1:]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Normalize the other columns using the scaler
normalized_cols = scaler.fit_transform(other_cols)

# Create a new DataFrame with the normalized columns
normalized_df = pd.DataFrame(normalized_cols, columns=other_cols.columns)

# Concatenate the first column with the normalized columns
result_df = pd.concat([first_col, normalized_df], axis=1)
result_df
```

```python
# Specify the file path and name
file_path = 'output.txt'

# Save the reshaped array to the file
# np.savetxt(file_path, X_normalized, fmt='%d', delimiter='\t')
np.savetxt(file_path, result_df)
```
